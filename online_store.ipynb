### Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

```python
df_pop = df\
    .query('year_of_release >= 2014')\
    .pivot_table(index = 'platform', values = 'all_sales', aggfunc = 'sum')\
    .reset_index()\
    .sort_values('all_sales',ascending = True)
df_pop
```

```python
height = df_pop['all_sales']
bars = df_pop['platform']

 
fig, ax = plt.subplots()


fig.set_size_inches(10,10)

# Create horizontal bars
ax.barh(y=bars, width=height, color=sns.color_palette('plasma', len(bars))[::-1]);

# Add title
ax.set_title('Продажи по платформам');
```

```python
play_station4 = df.query('year_of_release>=2014')[df['platform']=='PS4']
play_station4
```

```python
plt.figure(figsize=(17, 6))
ax = sns.scatterplot(x='user_score', y='all_sales', data=play_station4)



plt.title('Корреляция между оценками и продажами по платформе PS4', fontsize=15)
plt.xlabel('Оценка пользователя', fontsize=12)
plt.ylabel('Глобальные продажи',fontsize=12)
```

```python
play_station4['user_score'].corr(play_station4['all_sales'])
```

```python
plt.figure(figsize=(17, 6));
ax = sns.scatterplot(x='critic_score', y='all_sales', data=play_station4);



plt.title('Корреляция между оценками и продажами по платформе PS4', fontsize=15);
plt.xlabel('Оценка критика', fontsize=12);
plt.ylabel('Глобальные продажи',fontsize=12);
```

```python
play_station4['critic_score'].corr(play_station4['all_sales'])
```

- Имея следующую  градацию силы связи

     0 – полное отсутствие связи;
     
     0 – 0.3 – очень слабая;
     
     0.3 – 0.5 – слабая;
     
     0.5 – 0.7 – средняя;
     
     0.7 – 0.9 – высокая;
     
     0.9 – 1 – очень высокая.
     
  Можно заключить, что корреляция между оценками пользователей и продажами (-0.04) является очень слабой, в то время как корреляция продаж от оценок критиков (0.40) является просто слабой.

```python
for i, platform_name in enumerate(['PS4', 'XOne', '3DS', 'PS3', 'X360', 'WiiU']):
    user_cor = abs(platforms[i]['user_score'].corr(platforms[i]['all_sales']))
    crit_cor = abs(platforms[i]['critic_score'].corr(platforms[i]['all_sales']))
    if 0 <= user_cor <= 0.3 :
        print('Корреляция между отзывами пользователей и продажами для', platform_name, 'очень слабая :', user_cor)
        
    if 0.3 <= user_cor <= 0.5:
        print('Корреляция между отзывами пользователей и продажами для', platform_name, ' слабая :', user_cor)
    if 0.5 <= user_cor <= 0.7:
        print('Корреляция между отзывами пользователей и продажами для', platform_name, 'средняя :', user_cor)
    if 0.7 <= user_cor <= 0.9:
        print('Корреляция между отзывами пользователей и продажами для', platform_name, 'высокая :', user_cor)
    if 9 <= user_cor <= 1:
        print('Корреляция между отзывами пользователей и продажами для', platform_name, 'очень высокая :', user_cor)
    if 0 <= crit_cor <= 0.3:
        print('Корреляция между отзывами критиков и продажами для', platform_name, 'очень слабая :', crit_cor)
    if 0.3 <= crit_cor <= 0.5:
        print('Корреляция между отзывами  критиков и продажами для', platform_name, 'слабая :', crit_cor)
    if 0.5 <= crit_cor <= 0.7:
        print('Корреляция между отзывами  критиков и продажами для', platform_name, 'средняя :', crit_cor)
    if 0.7 <= crit_cor <= 0.9:
        print('Корреляция между отзывами  критиков и продажами для', platform_name, 'высокая :', crit_cor)
    if 9 <= crit_cor <= 1:
        print('Корреляция между отзывами  критиков и продажами для', platform_name, 'очень высокая :', crit_cor)
    print('\n\n')
```

- Исходя из этих данных, можно сделать вывод, что наибольшая корреляция между отзывами и продажами наблюдается у платформ PS3, X360 и WiiU в случае отзывов критиков, при этом она всё равно остается относительно слабой. Корреляция между отзывами пользователей и продажами в большинстве случаев является очень слабой, что может указывать на то, что отзывы не сильно влияют на успешность игр на этих платформах.
